```python
# 그래프 확인은 위하여 matplotlib 라이브러리
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 공부한 시간(x)와 점수(y)를 넘파이 배열로 선언
x = np.array([2,4,6,8])
y = np.array([81, 93, 91, 97])
```

```python
# 데이터의 분포를 점 그래프로 나타냅니다.
plt.scatter(x, y)
plt.show()
```

```python
# 기울기 a와 y절편 b 초기화
a = 0
b = 0
```

```python
# 학습률을 정합니다.
lr = 0.01
```

```python
# 몇 번 반복될지 설정합니다.
epochs =1
```

```python
# x 값이 총 몇 개인지 셉니다.
n = len(x)
```

```python
# 경사 하강법
for i in range(epochs):
  y_pred = a * x + b # 예측 값을 구하는 식입니다.
  error = y - y_pred # 실제 값과 비교한 오차를 error로 놓습니다.

  a_diff = (2/n) * sum(-x * (error)) # 오차 함수를 a로 편미분한 값입니다.
  b_diff = (2/n) * sum(-(error)) # 오차 함수를 b로 편미분한 값입니다.

  a = a - lr * a_diff # 학습률을 곱해 기존의 a 값을 업데이트
  b = b - lr * b_diff # 학습률을 곱해 기존의 b 값을 업데이트

  if i % 100 == 0 : # 100번 반복될 때마다 현재의 a 값, b 값을 출력합니다.
    print(f"학습률={lr}, epoch={i}, 기울기={a}, 절편={b}")  
```

```python
# 앞서 구한 최종 a 값을 기울기, b 값을 y 절편에 대입해 그래프를 그립니다.
y_pred = a * x + b
```

```python
# 그래프 출력
plt.scatter(x, y)
plt.plot(x, y_pred,'r')
plt.show()
```
